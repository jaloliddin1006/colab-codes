# With Polars

```python
import polars as pl

org_df = pl.DataFrame(orgs)
report_df = pl.DataFrame(reports)

# DataFrame-larni org_id va id ustunlari orqali birlashtirish
merged_df = report_df.join(org_df, left_on='org_id', right_on='id', how='left') # Use join for Polars DataFrames
# merged_df
```

```python
# barcha reportlarni period_id va form_type_id bo'yicha filter qilish
filtered_df = merged_df.filter((merged_df['period_id']==44) | (merged_df['form_type_id']==61))
filtered_df
```

```python
## har bitta parent saoto bo'yicha umumiy repostlar sonini chiqarish

df = merged_df.with_columns(
    pl.col('saoto').str.slice(0, 4).alias('parent_saoto')
)
result = df.group_by('period_id', 'form_type_id', 'status').agg(reports=pl.col('saoto').count())
# filter_df = df.filter((df['period_id']==44) & (df['form_type_id']==61) ).group_by('period_id', 'form_type_id', 'status', 'parent_saoto').agg(reports=pl.col('saoto').count())
filter_df = df.filter(df['period_id']==44).group_by('period_id', 'form_type_id', 'status', 'parent_saoto').agg(reports=pl.col('saoto').count())
new = filter_df.sort('parent_saoto', 'form_type_id')
new
```

```python
selected_df = new.select(["period_id", "form_type_id", "status", "reports", "parent_saoto"])
# selected_df
# Step 3: Aggregate and Pivot the DataFrame
pivoted_df = (
    selected_df
    .group_by(["period_id", "form_type_id", "status", "parent_saoto"])
    .agg(pl.sum("reports").alias("report_all"))  # Sum report counts
    .pivot(index=["period_id", "form_type_id", "status"], on="parent_saoto", values="report_all")
)


# # Handle missing values if needed
# pivoted_df = pivoted_df.fill_null(0)  # Replace missing values with 0 or another value as needed
# pivoted_df = pivoted_df.sort(["period_id", "form_type_id"])
# print(pivoted_df)



# Get a list of the 'saoto' columns and sort them
saoto_cols = sorted([col for col in pivoted_df.columns if col not in ["period_id", "form_type_id", "status"]])

# Create a new list of column names in the desired order
ordered_cols = ["period_id", "form_type_id", "status"] + saoto_cols

# Select the columns in the specified order
pivoted_df = pivoted_df.select(ordered_cols)
pivoted_df = pivoted_df.sort(["period_id", "form_type_id", 'status'])
print(pivoted_df)

pivoted_df.write_excel("result2.xlsx")
```
